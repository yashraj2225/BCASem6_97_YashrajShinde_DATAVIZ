```python
# Import Libraries and Load datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
iris = pd.read_csv(url, names=columns)

# Preview the data
iris.head()
```

```python
# Generate random array of 50 integers
data = np.random.randint(1, 101, 50)  # integers from 1 to 100

# Line Chart
plt.figure(figsize=(8,4))
plt.plot(data, marker='o', color='blue', linestyle='--')
plt.title('Line Chart of Random Integers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Scatter Plot
plt.figure(figsize=(8,4))
plt.scatter(range(len(data)), data, color='red')
plt.title('Scatter Plot of Random Integers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

# Histogram
plt.figure(figsize=(8,4))
plt.hist(data, bins=10, color='green', edgecolor='black')
plt.title('Histogram of Random Integers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(6,4))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title('Box Plot of Random Integers')
plt.show()
```

```python
# Add Two Outliers and display Box Plots
data_with_outliers = np.append(data, [150, 200])  # Adding two outliers

plt.figure(figsize=(6,4))
plt.boxplot(data_with_outliers, patch_artist=True, boxprops=dict(facecolor='pink'))
plt.title('Box Plot with Outliers')
plt.show()
```

```python
# c. Subject Names and Marks – Pie Chart and Bar Chart
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
marks = [85, 90, 78, 92, 88]

# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(marks, labels=subjects, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Marks Distribution in Subjects')
plt.show()

# Bar Chart
plt.figure(figsize=(8,4))
plt.bar(subjects, marks, color='skyblue', edgecolor='black')
plt.title('Marks in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
#d. Bar Plot – Frequency of Iris Species
species_count = iris['Species'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=species_count.index, y=species_count.values, palette='viridis')
plt.title('Frequency of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# pie Plots
plt.figure(figsize=(6,6))
plt.pie(species_count.values, labels=species_count.index, autopct='%1.1f%%', colors=sns.color_palette('bright'))
plt.title('Pie Chart of Iris Species Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='PetalLength', y='PetalWidth', hue='Species', data=iris, palette='deep')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Compare Sepal Length and Sepal Width
plt.figure(figsize=(6,4))
sns.scatterplot(x='SepalLength', y='SepalWidth', hue='Species', data=iris, palette='Set2')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

# Compare Petal Length and Petal Width
plt.figure(figsize=(6,4))
sns.scatterplot(x='PetalLength', y='PetalWidth', hue='Species', data=iris, palette='Set1')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
plt.figure(figsize=(12,8))
features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']

for i, feature in enumerate(features):
    plt.subplot(2,2,i+1)
    sns.boxplot(x='Species', y=feature, data=iris, palette='Set3')
    plt.title(f'{feature} Distribution by Species')

plt.tight_layout()
plt.show()
```
