# Continue with above dataset, find number of records for each distinct value of class attribute. Consider entire dataset and not the samples. Display column-wise mean, and median for iris dataset from (Hint: Use mean() and median() functions of pandas dataframe 

```python
import pandas as pd
data=pd.read_csv("iris.csv")
data
```

```python
print("For Sepal Length\nCount value is:  ",data["sepal_length"].count())
print("For Sepal width\nCount value is:  ",data["sepal_width"].count())
print("For petal Length\nCount value is:  ",data["petal_length"].count())
print("For petal width\nCount value is:  ",data["petal_width"].count())
```

```python
print("For sepal Length\nMean : ",data["sepal_length"].mean(),"\nMedian: ",data["sepal_length"].median())
print("\nFor sepal width\nMean : ",data["sepal_width"].mean(),"\nMedian: ",data["sepal_width"].median())
print("\nFor petal Length\nMean : ",data["petal_length"].mean(),"\nMedian: ",data["petal_length"].median())
print("\nFor petal width\nMean : ",data["petal_width"].mean(),"\nMedian: ",data["petal_width"].median())
```
