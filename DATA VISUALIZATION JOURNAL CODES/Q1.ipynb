```python
import pandas as pd
data={"Name":["parth","Dev","Sourav","Prathmesh","Nishant"],
      "Graduation Percentage":[89,90,99,78,77],
      "Age":[20,21,22,19,21]}

df=pd.DataFrame(data)
print(df)
print("Average age",df["Age"].mean())
print("Graduation Percentage: ",df["Graduation Percentage"].mean())

print("Median=",df["Graduation Percentage"].median())
print("Median=",df["Age"].median())

print("Mode=",df["Graduation Percentage"].mode())
print("Mode=",df["Age"].mode())

print("Range=",df["Graduation Percentage"].max()-df["Graduation Percentage"].min())
print("Range=",df["Age"].max()-df["Age"].min())

print("Quartile=",df["Graduation Percentage"].quantile(0.25))
print("Quartile=",df["Age"].quantile([0.25,0.50,0.75]))

print("Variance=",df["Graduation Percentage"].var())
print("Variance=",df["Age"].var())

```
